```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.arange(10)
print(x)
```

```python
x_square = x**2
x_cube = x**3
print(x_square)
print(x_cube)
```

```python
plt.plot(x, color="green")
plt.show()
```

```python
plt.plot(x, x_square, color="orange", label="square of x")
plt.plot(x, x_cube, color="green", label="cube of x")
plt.legend()
plt.xlabel("X")
plt.ylabel("values")
plt.title("$X^2$ vs $X^3$")
plt.show()
```

```python
# Scatter Plot
plt.scatter(x, x_square,marker="^", label="$x^2$", color="red")
plt.legend()
plt.xlabel("x")
plt.ylabel("x square")
plt.show()
```

```python
# Bar Charts
values = np.random.randint(0,10,8)*15
indices = np.arange(8)

print(values)
print(indices)
```

```python
plt.bar(indices, values)
plt.show()
```

```python
# Adding more bars in the same plot
values2 = np.random.randint(0,10,8)
indices2 = np.arange(8)

plt.bar(indices, values, color="red", label="red_label", width=0.3)
plt.bar(indices2+0.3,values2, color="blue", label="blue_label", width=0.3)
plt.legend()
plt.show()
```

```python
# Creating a pie chart
labels = ["AI","DL","DS","ML"]
counts = [40,80,60,320]
plt.pie(counts, labels=labels, shadow="True")
plt.show()
```

```python
plt.pie(counts, labels=labels, shadow="True", explode=(0,0,0,1))
plt.show()
```

```python
plt.pie(counts, labels=labels, colors=['pink','purple','blue','black'], shadow="True", explode=(0,1,1,0))
# plt.pie?
plt.show()
```
