```python
import requests
from bs4 import BeautifulSoup
```

```python
BASE_URL = "https://www.goodreads.com/quotes?page="
```

```python
quotes = []
```

```python
def parse(r):
    soup = BeautifulSoup(r.text)
    div_quotes = soup.find('div', attrs={'class': 'quotes'})
    
    for div_quote in div_quotes.find_all('div', attrs={'class': 'quote'}):
        
        quote_text    = div_quote.select_one('div.quoteText')
        quote_footer  = div_quote.select_one('div.quoteFooter')
        
#         # Decomposing script 'child'
#         while True:
#             try:
#                 quote_text.script.decompose()
#             except:
#                 print('!')
#                 break
        
        # CLEANING, EXTRACTING
        quote, author = quote_text.text.strip().split('―')
        footer   = quote_footer.text.strip().split('\n')
        tags = footer[1:-1]
        likes = footer[-1]
        
        quotes_dict = {}
        quotes_dict['quote'] = quote.strip()[1:-1]
        quotes_dict['author'] = author.split(',')[0].strip()
        quotes_dict['tags'] = [t.strip().replace(',', '') for t in tags if t]
        quotes_dict['likes'] = likes.strip().replace(' likes', '')
        quotes.append(quotes_dict)
    
    

def scrape(n_pages=10):
    for p in range(n_pages):
        print("Scraping Page #{}".format(p+1))
        url = BASE_URL + str(p+1)
        
        try:
            r = requests.get(url)
            r.raise_for_status()
        except Exception as e:
#         except requests.ConnectionError:
            print("Error Occurred: {}".format(e))
    
    parse(r)
    
        
```

```python
scrape()
```

## Writing Data to File

```python
import json
with open('quotes.json', 'w') as f:
    json.dump(quotes, f, indent=4)
```

```python
import csv
with open('quotes.csv', 'w') as f:
    writer = csv.DictWriter(f, fieldnames=["quote", "author", "tags", "likes"])
    writer.writeheader()
    for quote in quotes:
        writer.writerow(quote)
```
